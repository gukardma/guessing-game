```python
import csv
import sqlite3
import argparse
import os
import sys
import pandas as pd
```

```python
proj_dir = os.path.abspath(os.path.join('.'))
```

```python
df = pd.read_csv(os.path.join(proj_dir, 'zoo2.csv'))
```

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("data.db")

# 3. Write the DataFrame into a table
df.to_sql(
    name="animals",      # table name inside SQLite
    con=conn,             # or con=engine
    if_exists="replace",  # 'fail' | 'replace' | 'append'
    index=False           # drop the pandas index column
)

# 4. (Optional) verify
print(pd.read_sql("SELECT * FROM animals", conn))

# 5. Always close connections
conn.close()
```
